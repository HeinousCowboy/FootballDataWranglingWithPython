```python
df = df.drop_duplicates()
```

```python
df["Team & Contract"] = df["Team & Contract"].str.strip("\n")
```

```python
df
```

```python
df= df.rename(columns={'Team & Contract': 'Team',
'LongName':  'Player Name','↓OVA':  'Overall Rating'})
```

```python
df[["Team","Contract"]]= df['Team'].str.split("\n",1,expand=True)
```

```python
df
```

```python
df["Weight"]=df["Weight"].str.replace("lbs","")
```

```python
df["Height"]=df["Height"].str.replace("'",".")
df["Height"]=df["Height"].str.replace('"', "")
```

```python
df
```

```python
df[['Start of Contract','End of Contract']]=df['Contract'].str.split("~", 1, expand=True)
```

```python
df["Height"] = df["Height"].astype(float)
df["Weight"] = df["Weight"].astype(int)
```
